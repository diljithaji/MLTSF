## Implementing Moving Average

```python
#Function for Moving Average

def calculate_moving_average(data, k):
    moving_averages = []
    for i in range(len(data) - k + 1):
        window = data[i : i + k]
        window_average = sum(window) / k
        moving_averages.append(window_average)
    return moving_averages
```

```python
# Batting avg for few games
batting_averages = [
    0.280, 0.300, 0.295, 0.310, 0.320, 0.315, 0.330, 0.325, 0.340, 0.335,
    0.350, 0.345, 0.355, 0.360, 0.358, 0.362, 0.365, 0.370, 0.368, 0.372,
    0.380, 0.375, 0.385, 0.390, 0.388, 0.392, 0.395, 0.400
]

"""
taking a window size of 10.
taking 10 consecutive games at a time
"""
window_size = 10

# 1st Order moving avg
first_order = calculate_moving_average(batting_averages, window_size)

# 2nd Order moving avg (based on first order result)
second_order = calculate_moving_average(first_order, window_size)

print("Original Batting Averages:", batting_averages)
print("\n")
print("------ 1st Order Moving Average ------")
for i, ma in enumerate(first_order,start=window_size):
    print(f"Game {i}: 10-Game 1st Order MA= {ma:.3f}")

print("\n------ 2nd Order Moving Average ------")
for i, ma in enumerate(second_order, start=2*window_size-1):
    print(f"Game {i}: 10-Game 2nd Order MA= {ma:.3f}")
```

```python
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(10,6))

# Plot original batting average
plt.plot(range(1,len(batting_averages)+1),batting_averages,marker='o',label="Original Batting Average")

# Plot 1st order Moing avg
plt.plot(range(window_size,len(batting_averages)+1),first_order,marker='s',label=f"1st Order MA ({window_size}-game)")

# Plot 2nd order Moving avg
plt.plot(range(2*window_size-1,len(batting_averages)+1),second_order,marker='^',label=f"2nd Order MA ({window_size}-game)")

plt.xlabel("Game Number")
plt.ylabel("Batting Average")
plt.title(f"Baseball Batting Average with {window_size}-Game Moving Averages")
plt.legend()
plt.grid(True)
plt.show()
```
